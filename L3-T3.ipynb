```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Dataset.csv")
print(df.head())
```

```python
x1=df['Has Online delivery']
y=df['Average Cost for two']
x2=df['Has Table booking']
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1 , y=y)
plt.xlabel('Online delivery')
plt.ylabel('Average Cost for two')
plt.title('Online delivery affected by cost')

plt.subplot(1,2,2)
sns.boxplot(x=x2 , y=y)
plt.xlabel('Table booking')
plt.ylabel('Average Cost for two')
plt.title('Table booking affected by cost')
plt.tight_layout()
plt.show()
```

```python
print("Both the online delivery and Table booking are negatively impact by the cost")
```

```python
print("Hih priced Resturants which have average cost more than 10,000 did not provide Online delivery as well as Table booking facilities.")      
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1 , y='Average Cost for two', data=df.loc[df['Average Cost for two']<10000])
plt.title('Online delivery by cost')

plt.subplot(1,2,2)
sns.boxplot(x=x2 , y='Average Cost for two', data=df.loc[df['Average Cost for two']<10000])
plt.title('Table booking by cost')
plt.tight_layout()
plt.show()
```

```python
print("The Online delivery facility is present for only those resturant,where the average cost is less than 1000.")
print("The Table booking facility is available in resturants with an average cost ranging from 1000 to 2000.")
```
